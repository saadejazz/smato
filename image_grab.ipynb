```python
from icrawler.builtin import GoogleImageCrawler

google_crawler = GoogleImageCrawler(
    feeder_threads=1,
    parser_threads=2,
    downloader_threads=4,
    storage = {
        'root_dir': 'downloaded_images'
    })

filters = dict(
    date=((2022, 1, 1), (2022, 12, 30)))

google_crawler.crawl(keyword = 'pedestrians with phone', max_num = 1000, file_idx_offset = 0)
```

```python
from bing_image_downloader import downloader

downloader.download("pedestrians with phone", limit = 1000,  output_dir = 'bing_images', adult_filter_off=True, force_replace=False, timeout=60, verbose=True)
```
